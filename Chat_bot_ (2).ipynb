```python
import os
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import HumanMessage, AIMessage
from dotenv import load_dotenv

load_dotenv()
gemini_key = os.getenv("GEMINI_API_KEY")
os.environ["GOOGLE_API_KEY"] = gemini_key

chat_model = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
chat_history = []

# Save in the same folder as this notebook/script
log_file = os.path.join(os.getcwd(), "chat_history.txt")

print(f"Chat history will be saved to: {log_file}")

while True:
    user_input = input("You: ")
    if user_input.lower() in ["exit", "quit"]:
        print("Bot: Goodbye!")
        break

    chat_history.append(HumanMessage(content=user_input))
    response = chat_model.invoke(chat_history)
    chat_history.append(response)
    print("Bot:", response.content)

    # Save to file and force flush
    with open(log_file, "a", encoding="utf-8") as f:
        f.write(f"You: {user_input}\n")
        f.write(f"Bot: {response.content}\n\n")
        f.flush()  # ensures messages are immediately written to chat_history.txt
```
